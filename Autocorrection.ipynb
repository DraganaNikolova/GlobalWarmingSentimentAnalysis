```python
from nltk.tokenize import word_tokenize
import re
import pandas as pd
df = pd.read_pickle("preprocessed_unlabeled.pkl")

# Make corpus
corpus = {}
for text in df['clean']:
    for i in word_tokenize(text):
        i = re.sub("[^a-zA-Z]","",i)  
        if len(i) > 1:
            corpus[i] = ''

# Extract non words
import enchant
d = enchant.Dict("en_US")

non_words = []
for word in corpus.keys():
    if len(word) > 3:
        if d.check(word) is not True:
            non_words.append(word)
```

```python
from autocorrect import Speller
spell = Speller(lang='en')

# Save autocorrected words
corrected = {}
for word in non_words:
    c = spell(word)
    if d.check(c) is True:
        corrected[word] = c
        corpus[word] = c
corrected
```

```python
# Save autocorrections
import pickle

# Save dictionary to pkl file
with open('autocorrections.pkl', 'wb') as fp:
    pickle.dump(corrected, fp)
    print('dictionary saved successfully to file')
```

```python
# Load autocorrections
with open('autocorrections.pkl', 'rb') as fp:
    autocorrections = pickle.load(fp)
    print(autocorrections)
```
